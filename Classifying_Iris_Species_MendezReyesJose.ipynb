```
# Its function is similar to that of a dictionary since it can contain values.
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```
# That is a simple description of the data that is gathered
print("Keys of iris_dataset: \n{}".format(iris_dataset.keys()))
```

```
# the rows in the matrix represent the data types and the columns represent the measures that are 4.
print("Type of data: {}".format(type(iris_dataset['data'])))
```

```
# In the matrix you can save an infinity of types, something, dogs, their breeds, etc. and here the values of the first 5 samples that were used are shown:
print("Shape of data: {}".format(iris_dataset['data'].shape))
```

```
# The outputs will be designated in the following nomenclature:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```
# In the output of train_test_split, the function X_train, X_test, y_train and y_test, which are all NumPy arrays. X_train contains 75% of the rows and X_test has 25%:
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

```

print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

```
# The points are lit depending on the data types, in that case flowers, from the iris first the NumPy Matrix is converted into a pandas DataFrame, pandas also has a function for creating parcels called scatter_matrix.
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets

iris_dataset = datasets.load_iris()
X = iris_dataset.data
Y = iris_dataset.target

iris_dataframe = pd.DataFrame(X, columns=iris_dataset.feature_names)

# Create a scatter matrix from the dataframe, color by y_train
grr = pd.plotting.scatter_matrix(iris_dataframe, c=Y, figsize=(8, 8), marker='o',
                                 hist_kwds={'bins': 20}, s=60, alpha=.8)
```

```
# We are going to introduce KNeighborsClassifier in the neighbors module to create an instance of the class in an object and the most important parameter is the number of neighbors, and it will be set to 1:
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```
# To build the model in the training set, the settings are used in the knn mode, which is based on talking about arguments in the NumPy X_train matrix because it has training data and the NumPy y_train matrix of the training labels.
knn.fit(X_train, y_train)
```

```
# scikit-learn printing can cause long strings but don't pay much attention or worry about them, the output will not be shown as it has no relevant information.
import numpy as np

X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape: {}".format(X_new.shape))
```

```
# Now we can make predictions with these data that we could not, if a type of flower were found with certain measurements how we could know what type of flower it is, putting this data in a Numpy matrix, once again to calculate its shape: multiply the number of samples by the one of characteristics.
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted target name: {}".format(
iris_dataset['target_names'][prediction]))
```

```
# Only in this flower were the measurements made in a two-dimensional matrix because scikit-learn always expects two-dimensional.
y_pred = knn.predict(X_test)
print("Test set predictions:\n {}".format(y_pred))
```

```
# To make a prediction we will call the knn object method
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```

```
# Later, we can create a prediction for each iris in a data test and then buy to check and see how effective the method is for the predictions and which species were predicted.
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```

```
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
